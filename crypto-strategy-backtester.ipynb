```python
from src.regime import add_market_regime

df = load_crypto_data("BTC-USD", start="2021-01-01")
df = add_indicators(df, short_window=20, long_window=50)
df = add_market_regime(df)
df = ma_crossover_strategy(df)

results = run_backtest(
    df,
    initial_capital=10000,
    transaction_cost=0.001,
    slippage=0.001,
    stop_loss_atr=2,
    take_profit_atr=3,
    risk_per_trade=0.01
)

results[["Date", "Close", "trend_regime", "volatility_regime", "strategy_return"]].head()
```

```python
regime_summary = results.groupby("trend_regime")["strategy_return"].agg(
    total_return=lambda x: (1 + x).prod() - 1,
    avg_return="mean",
    volatility="std",
    count="count"
)

regime_summary
```

```python
vol_regime_summary = results.groupby("volatility_regime")["strategy_return"].agg(
    total_return=lambda x: (1 + x).prod() - 1,
    avg_return="mean",
    volatility="std",
    count="count"
)

vol_regime_summary
```
